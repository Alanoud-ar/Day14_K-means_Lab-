# EDA

## Create a scatterplot of Grad.Rate versus Room.Board where the points are colored by the Private column. 

```python
plt.figure(figsize=(15,8))
sns.relplot(
    x="Room.Board", 
    y="Grad.Rate", 
    hue="Private",
#     size="weight",
    alpha=.8, 
    palette="muted",
    height=6, 
    data=df
);
plt.grid()
```

## Create a scatterplot of F.Undergrad versus Outstate where the points are colored by the Private column.

```python
sns.relplot(
    x="Outstate", 
    y="F.Undergrad", 
    hue="Private",
#     size="weight",
    alpha=.8, 
    palette="muted",
    height=6, 
    data=df
);
plt.grid()
```

## Use sns.FacetGrid to create a stacked histogram showing Out of State Tuition based on the Private column. 

```python
g = sns.FacetGrid(df, hue="Private", size=7, aspect=1.7)
g.map(plt.hist, "Outstate", bins=20, histtype='barstacked', stacked=True);
plt.show()
g.grid()
```

## Use sns.FacetGrid to create a stacked histogram showing Grad.Rate based on the Private column. 

```python

g = sns.FacetGrid(df, hue="Private", size=7, aspect=1.7)
g.map(plt.hist, "Grad.Rate", bins=20, histtype='barstacked', stacked=True);
plt.show()
```

## K Means Model

```python
from sklearn.cluster import KMeans
```

## Fit the model to all the data except for the Private label.

```python
x = df.iloc[:,1:]
kmeans.fit(x)
```
